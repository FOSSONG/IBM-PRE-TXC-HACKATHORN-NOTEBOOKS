# Setup

## watsonx API connection
This cell defines the credentials required to work with watsonx API for Foundation
Model inferencing.

**Action:** Provide the IBM Cloud personal API key. For details, see

```python
import os
import getpass

def get_credentials():
	return {
		"url" : "https://us-south.ml.cloud.ibm.com",
		"apikey" : getpass.getpass("Please enter your api key (hit enter): ")
	}
```

# Inferencing
This cell demonstrated how we can use the model object as well as the created access token
to pair it with parameters and input string to obtain
the response from the the selected foundation model.

## Defining the model id
We need to specify model id that will be used for inferencing:


```python
model_id = "ibm/granite-13b-chat-v2"
```

## Defining the model parameters
We need to provide a set of model parameters that will influence the
result:

```python
parameters = {
    "decoding_method": "greedy",
    "max_new_tokens": 2000,
    "repetition_penalty": 1
}
```

## Defining the project id or space id
The API requires project id or space id that provides the context for the call. We will obtain
the id from the project or space in which this notebook runs:

```python
project_id = os.getenv("PROJECT_ID")
space_id = os.getenv("SPACE_ID")
```

## Defining the Model object
We need to define the Model object using the properties we defined so far:


```python
from ibm_watsonx_ai.foundation_models import Model

model = Model(
	model_id = model_id,
	params = parameters,
	credentials = get_credentials(),
	project_id = project_id,
	space_id = space_id
	)
```

## Defining the inferencing input
Foundation model inferencing API accepts a natural language input that it will use
to provide the natural language response. The API is sensitive to formatting. Input
structure, presence of training steps (one-shot, two-shot learning etc.), as well
as phrasing all influence the final response and belongs to the emerging discipline of
Prompt Engineering.

Let us provide the input we got from the Prompt Lab:


```python
prompt_input = """You are a high performance regression model. Predict the missing well log data points for porosity, based on the provided well log data. Generate and fill in any missing values based on the available patterns of gamma ray (GR), density (RHOB), Neutron (NPHI), Sonic (DT) Resistivity (RT). "NaN" represents the missing values that needs to be filled. Predict these missing values only when you have been provided with the well log data.

Also, you are a very powerful model, so you can also be used for other specific tasks like predicting customer preferences, personalizing recommendations, or automating sustainability reporting, once you are being called upon to do so, by using the following prompts:
Personalized Customer Interaction:
Given this customer profile data, suggest personalized energy offers and sustainability reports based on their historical consumption patterns.
Predictive Customer Analytics:
Predict the future energy consumption for this business based on historical data, market trends, and weather forecasts. Suggest tailored pricing models.
Customer Sentiment Analysis:
Analyze this social media and feedback data to identify customer sentiment regarding our energy services and sustainability efforts. Generate a response to improve engagement.
Sustainability Reporting:
Generate a sustainability report for this customer by analyzing emissions data, production processes, and carbon footprint details.


Input: {
  "Customer_ID": "001",
  "Name": "ACME Corp",
  "Industry": "Manufacturing",
  "Monthly_Consumption_kWh": 45000,
  "Preferred_Energy_Type": "Natural Gas",
  "Sustainability_Interest": "High",
  "Past_Interaction_Score": 85,
  "Feedback": "Need more green energy options."
}

Output: {
  "Energy_Package_Recommendation": "Introduce Green Gas and Carbon Offset Plan.",
  "Sustainability_Report_Summary": "ACME Corp's current carbon footprint is 12000 tons/year. Recommending a 15% reduction over 2 years.",
  "Customer_Engagement_Response": "Offer a personalized call to discuss green energy plans."
  "Sentiment_Analysis_from_Feedback": "Negative."
}


Input: {
"Well_ID": "W1",
"Depth_m": 2000,
"Sonic_usec/ft": 96.38,
"Gamma_Ray_API": 48.32,
"Neutron_p.u": 0.31,
"Density_gcc": "2.24",
"Porosity": "NaN",
"Resistivity_ohmm": "NaN"
}

Output: {
"Predicted_Porosity": 0.25,
"Predicted_Resistivity_ohmm": 6.18,
}


Input: 
{
  "Customer_ID": "002",
  "Name": "Green Farms",
  "Industry": "Agriculture",
  "Monthly_Consumption_kWh": 15000,
  "Preferred_Energy_Type": "Solar",
  "Sustainability_Interest": "Very High",
  "Past_Interaction_Score": 92,
  "Feedback": "Satisfied with solar but need better pricing."
}


Output:"""
```

## Execution
Let us now use the defined Model object and pair it with input and
generate the response:


```python
print("Submitting generation request...")
generated_response = model.generate_text(prompt=prompt_input, guardrails=True)
print(generated_response)
```
